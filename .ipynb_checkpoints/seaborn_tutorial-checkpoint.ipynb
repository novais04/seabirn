# Visualizações Relacionadas a Estatistica

## Relacionando variáveis com gráficos de dispersão

O __gráfico de dispersão__ é um dos pilares da visualização estatística. Representa a distribuição conjunta de duas variáveis usando uma nuvem de pontos, onde cada ponto representa uma observação no conjunto de dados. Esta representação permite ao olho inferir uma quantidade substancial de informações sobre se existe alguma relação significativa entre eles.

Existem várias maneiras de desenhar um gráfico de dispersão(scatter plot) . A mais básica, que deve ser usada quando ambas as variáveis são numéricas, é a função ```scatterplot()```. No tutorial de visualização categórica, veremos ferramentas especializadas para usar gráficos de dispersão (scatterplot) para visualizar dados categóricos. O __scatterplot()__ é o tipo padrão em __relplot()__ (também pode ser forçado configurando ```kind="scatter"```):

````python
# importando Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='darkgrid')

import warnings
warnings.filterwarnings('ignore')
````

### Dataset tips

__contexto:__Um garçom registrou informações sobre cada gorjeta que recebeu durante alguns meses trabalhando em um restaurante. Ao todo ele registrou 244 gorgetas.

* __total_bill__:  Valor da conta (dólares);
* __tip__: O valor da gorjeta (dólares);
* __sex__: Genero do pagador (neste caso femino ou masculino);
* __smoker__: Se a festa incluia fumantes;
* __day__: O dia da semana (quinta(Thur, sexta (Fri), sabado (Sat) ou domingo (Sun));
* __time__: Almoço ou Janta;
* __size__: Quantidade de pessoas na mesa

````python
tips = sns.load_dataset('tips') # importanto o dataset tips (gordetas de carços US)
tips.sample(3)
````

````python
sns.relplot(data=tips, x='total_bill', y='tip');
````

__Observação:__ Embora os pontos sejam plotados em duas dimensões, outra dimensão pode ser adicionada ao gráfico colorindo os pontos de acordo com uma terceira variável. Em seaborn, isso é chamado de ```“hue semantic”```, porque a cor do ponto ganha significado:

para muda a cor use o paramentro __hue__

````python
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker');
````

> Para enfatizar a diferença entre as classes e melhorar a acessibilidade, você pode usar um estilo de marcador diferente para cada classe: use __style__


````python
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', style='smoker')
````

> Também é possível representar quatro variáveis alterando o matiz e o estilo de cada ponto de forma independente. Mas isso deve ser feito com cuidado, pois o olho é muito menos sensível à forma do que à cor:

````python
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', style='time')
````

> Nos exemplos acima, a semântica de matiz era categórica, portanto a paleta qualitativa padrão foi aplicada. Se a semântica de matiz for numérica (especificamente, se puder ser convertida em flutuante), a coloração padrão mudará para uma paleta sequencial:

````python
sns.relplot(data=tips, x='total_bill', y='tip', hue='size' )
````

> Em ambos os casos, você pode personalizar a paleta de cores. Existem muitas opções para fazer isso. Aqui, personalizamos uma paleta sequencial usando a interface de string para ```cubehelix_palette()```:

````python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='size',
    palette='ch:r=-.5, l=.75'
)
````

> O terceiro tipo de variável semântica altera o tamanho de cada ponto:

````python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    size='size'
)
````

> Ao contrário de matplotlib.pyplot.scatter(), o valor literal da variável não é usado para escolher a área do ponto. Em vez disso, o intervalo de valores em unidades de dados é normalizado em um intervalo em unidades de área. Esta gama pode ser personalizada:

````python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    size='size',
    sizes=(15,200)
)
````

## Enfatizando a continuidade com gráficos de linhas

Os __gráficos de dispersão__ são altamente eficazes, mas não existe um tipo de visualização universalmente ideal. Em vez disso, a representação visual deve ser adaptada às especificidades do conjunto de dados e à pergunta que você está tentando responder com o gráfico.

Com alguns conjuntos de dados, você pode querer entender as mudanças em uma variável em função do tempo ou em uma variável contínua semelhante. Nessa situação, uma boa opção é traçar um __gráfico de linhas__. No Seaborn, isso pode ser feito pela função ```lineplot()```, diretamente ou com ```relplot()``` definindo __kind="line"__:

````python
# Obtendo dataser
dowjones=sns.load_dataset("dowjones")
sns.relplot(data=dowjones, x='Date', y='Price', kind='line')
````

## Agregação e representação de incerteza

Conjuntos de dados mais complexos terão múltiplas medições para o mesmo valor da variável x. O comportamento padrão no __seborn__ é agregar as múltiplas medições em cada valor de x traçando a média e o intervalo de confiança de 95% em torno da média:

### Dataset fmri

````python
fmri = sns.load_dataset('fmri')
fmri
````

````python
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    kind='line'
)
````

> Os intervalos de confiança são calculados usando bootstrapping, que pode consumir muito tempo para conjuntos de dados maiores. Portanto, é possível desativá-los

````python
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    kind='line',
    errorbar=None
)
````

> Outra boa opção, especialmente com dados maiores, é representar a dispersão da distribuição em cada ponto no tempo traçando o desvio padrão em vez de um intervalo de confiança:

````python
sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    errorbar='sd'
)
````

## Plotando subconjuntos de dados com mapeamentos semânticos

A função __lineplot()__ tem a mesma flexibilidade que __scatterplot()__: ela pode mostrar até três variáveis adicionais modificando o matiz, tamanho e estilo dos elementos do gráfico. Isso é feito usando a mesma API de __scatterplot()__, o que significa que não precisamos parar e pensar sobre os parâmetros que controlam a aparência das linhas versus pontos no matplotlib.

Usar a semântica em __lineplot()__ também determinará como os dados serão agregados. Por exemplo, adicionar uma semântica de matiz com dois níveis divide o gráfico em duas linhas e faixas de erro, colorindo cada uma para indicar a qual subconjunto de dados elas correspondem.

````python
sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    hue='event'
)
````

> Adicionar uma semântica de estilo a um gráfico de linha altera o padrão de traços na linha por padrão:

````python
sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    hue='region',
    style='event'
)
````

> Mas você pode identificar subconjuntos pelos marcadores usados em cada observação, seja junto com os traços ou em vez deles:

````python
sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    hue='region',
    style='event',
    kind='line',
    dashes=False,
    markers=True
)
````

> Tal como acontece com os gráficos de dispersão, tenha cuidado ao criar gráficos de linhas usando semântica múltipla. Embora às vezes informativos, eles também podem ser difíceis de analisar e interpretar. Mas mesmo quando você está examinando apenas alterações em uma variável adicional, pode ser útil alterar a cor e o estilo das linhas. Isso pode tornar o gráfico mais acessível quando impresso em preto e branco ou visualizado por alguém com daltonismo:

````python
sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    hue='event',
    style='event',
    kind='line'
)
````

> O mapa de cores padrão e o tratamento da legenda em lineplot() também dependem se a semântica de matiz é categórica ou numérica:

````python
dots = sns.load_dataset('dots').query("align=='dots'")
dots
````

````python
sns.relplot(
    data=dots,
    x='time',
    y='firing_rate',
    hue='coherence',
    style='choice',
    kind='line'
)
````

## Mostrando vários relacionamentos com facetas

Enfatizamos neste tutorial que, embora essas funções possam mostrar diversas variáveis semânticas ao mesmo tempo, nem sempre é eficaz fazê-lo. Mas e quando você quiser entender como um relacionamento entre duas variáveis depende de mais de uma outra variável?

A melhor abordagem pode ser fazer mais de um gráfico. Como `relplot() `é baseado no `FacetGrid`, isso é fácil de fazer. Para mostrar a influência de uma variável adicional, em vez de atribuí-la a um dos papéis semânticos do enredo, use-a para “facetar” a visualização. Isso significa que você cria vários eixos e plota subconjuntos de dados em cada um deles:




````python
sns.relplot(
#    kind='line',
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    col='time'  # cria um grafico para cada tipo de time
)
````

Você também pode mostrar a influência de duas variáveis desta forma: uma por lapidação nas colunas e outra por lapidação nas linhas. À medida que você começa a adicionar mais variáveis à grade, você pode querer diminuir o tamanho da figura. Lembre-se que o tamanho do `FacetGrid`` é parametrizado pela altura e proporção de cada faceta:

````python
sns.relplot(
    data=fmri,
    kind='line',
    x='timepoint',
    y='signal',
    hue='subject',
    col='region', # cria um gráfico para cada região (region)
    row='event',
    height=3,
    estimator=None
)
````

Quando você deseja examinar os efeitos em vários níveis de uma variável, pode ser uma boa ideia facetar essa variável nas colunas e depois “envolver” as facetas nas linhas:

````python
sns.relplot(
    kind='line',
    data=fmri.query("region=='frontal'"),
    x='timepoint',
    y='signal',
    hue='event',
    style='event',
    col='subject',
    col_wrap=5,  # quebra de colunas no caso aqui  o máximo de colunas será 5
    height=3,
    aspect=.8,
    linewidth=2.5
);
````

Essas visualizações, às vezes chamadas de gráficos de “reticulação” ou “múltiplos pequenos”, são muito eficazes porque apresentam os dados em um formato que facilita ao olho detectar padrões gerais e desvios desses padrões. Embora você deva aproveitar a flexibilidade proporcionada por `scatterplot()` e `relplot()`, sempre tente ter em mente que vários gráficos simples geralmente são mais eficazes do que um gráfico complexo.

# Visualizando Distribuições de <span style="color:blue">**Dados**</span>

<img src="../images/function_overview_8_0.png" style="height: 300px; width:300px;"/>

Um passo inicial em qualquer esforço para analisar ou modelar dados deve ser compreender como as variáveis são distribuídas. As técnicas de visualização da distribuição podem fornecer respostas rápidas para muitas questões importantes. Que faixa as observações cobrem? Qual é a sua tendência central? Eles estão fortemente inclinados em uma direção? Existem evidências de bimodalidade? Existem valores discrepantes significativos? As respostas a estas perguntas variam entre subconjuntos definidos por outras variáveis?

O módulo de distribuições contém diversas funções projetadas para responder perguntas como essas. As funções em nível de eixo são `histplot()`,` kdeplot()`,` ecdfplot()` e `rugplot()`. Eles são agrupados nas funções `displot()`, `jointplot()` e `pairplot()` em nível de figura.

Existem diversas abordagens diferentes para visualizar uma distribuição, e cada uma tem suas vantagens e desvantagens relativas. É importante compreender esses fatores para que você possa escolher a melhor abordagem para seu objetivo específico.


### Import Packeges

````python
# importando Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='darkgrid')

import warnings
warnings.filterwarnings('ignore')
````

## Plotting univariate histograms
Talvez a abordagem mais comum para visualizar uma distribuição seja o histograma. Esta é a abordagem padrão em `displot()`, que usa o mesmo código subjacente de `histplot()`. 

Um histograma é uma espécie de gráfico de barras que demonstra uma distribuição de frequências. No histograma, a base de cada uma das barras representa uma classe e a altura representa a quantidade ou frequência absoluta com que o valor de cada classe ocorre. Ao mesmo tempo, ele pode ser utilizado como um indicador de dispersão de processos.

### Dataset penguins

**Definição da coluna de dados dos pinguins**

__species__ um fator que denota espécies de pinguins (Adélie, Chinstrap e Gentoo)

__island__ um fator que denota uma ilha no Arquipélago Palmer, Antártica (Biscoe, Dream ou Torgersen)

__bill_length_mm__ um número que indica o comprimento da nota (milímetros)

__bill_profundidade_mm__ um número que indica a profundidade da nota (milímetros)

__flipper_length_mm__ um número inteiro que denota o comprimento do flipper (milímetros)

__body_mass_g um número inteiro__ denotando a massa corporal (gramas)

__sex__ um fator que denota o sexo do pinguim (feminino, masculino)

````python
#sns.displot?
````

````python
penguins = sns.load_dataset('penguins')
penguins.sample(5)
````

````python
sns.displot(penguins, x="flipper_length_mm")
````

Este gráfico fornece imediatamente alguns insights sobre a variável __flipper_length_mm__. Por exemplo, podemos ver que o comprimento mais comum da nadadeira é de cerca de 195 mm, mas a distribuição parece bimodal, portanto este número não representa bem os dados.

## Escolhendo o tamanho da caixa

O tamanho dos compartimentos é um parâmetro importante, e usar o tamanho incorreto do compartimento pode induzir em erro, ocultando características importantes dos dados ou criando características aparentes a partir da variabilidade aleatória. Por padrão, `displot()`/`histplot()` escolhe um tamanho de compartimento padrão com base na variação dos dados e no número de observações. Mas você não deve confiar demais nessas abordagens automáticas, porque elas dependem de suposições específicas sobre a estrutura dos seus dados. É sempre aconselhável verificar se suas impressões da distribuição são consistentes em diferentes tamanhos de compartimentos. Para escolher o tamanho diretamente, defina o parâmetro `binwidth`:

````python
penguins.columns
````

````python
sns.displot(
    penguins,
    x='flipper_length_mm',
    binwidth=3
)
````

>  Em outras circunstâncias, pode fazer mais sentido especificar o número de caixas, em vez do seu tamanho:    

````python
sns.displot(penguins, x='flipper_length_mm', bins=15)
````

Um exemplo de situação em que os padrões falham é quando a variável assume um número relativamente pequeno de valores inteiros. Nesse caso, a largura padrão do compartimento pode ser muito pequena, criando lacunas estranhas na distribuição:

````python
tips = sns.load_dataset("tips")
tips.sample(3)
````

````python
sns.displot(tips, x='size')
````

Um exemplo de situação em que os padrões falham é quando a variável assume um número relativamente pequeno de valores inteiros. Nesse caso, a largura padrão do compartimento pode ser muito pequena, criando lacunas estranhas na distribuição (`bin`)
:

````python
sns.displot(tips, x="size", bins=[1, 2, 3, 4, 5, 6, 7])
````

Isso também pode ser feito definindo `discrete = True`, que escolhe quebras de bin que representam os valores exclusivos em um conjunto de dados com barras centralizadas em seu valor correspondente.

````python
sns.displot(tips, x='size', discrete=True)
````

Também é possível visualizar a distribuição de uma variável categórica utilizando a lógica de um histograma. Os compartimentos discretos são definidos automaticamente para variáveis categóricas, mas também pode ser útil __“shrink”__ as barras ligeiramente para enfatizar a natureza categórica do eixo:

````python
sns.displot(tips, x='day', shrink=0.8)
````

## Condicionamento em outras variáveis

Depois de entender a distribuição de uma variável, a próxima etapa geralmente é perguntar se os recursos dessa distribuição diferem entre outras variáveis no conjunto de dados. Por exemplo, o que explica a distribuição bimodal dos comprimentos das nadadeiras que vimos acima? `displot()` e `histplot()` fornecem suporte para subconjuntos condicionais por meio da semântica de matiz. Atribuir uma variável a matiz desenhará um histograma separado para cada um de seus valores exclusivos e os distinguirá por cor:

````python
sns.displot(penguins, x='flipper_length_mm', hue='species')
````

Por padrão, os diferentes histogramas são “em camadas” uns sobre os outros e, em alguns casos, podem ser difíceis de distinguir. Uma opção é alterar a representação visual do histograma de um gráfico de barras para um gráfico de “passos”:

````python
sns.displot(penguins, x='flipper_length_mm', hue='species', element='step')
````

````python
sns.color_palette("YlGnBu",n_colors=10 )
````

````python
sns.color_palette("rocket",n_colors=10 )
````

> Lista de cores do sns.color_pallette
[color_pallette list name](https://r02b.github.io/seaborn_palettes/)


Alternativamente, em vez de colocar cada barra em camadas, elas podem ser “empilhadas” ou movidas verticalmente. Neste gráfico, o contorno do histograma completo corresponderá ao gráfico com apenas uma única variável:

````python
sns.displot(penguins, x="flipper_length_mm", hue="species", multiple="stack")
````

O histograma empilhado enfatiza a relação parte-todo entre as variáveis, mas pode obscurecer outras características (por exemplo, é difícil determinar a moda da distribuição de Adelie. Outra opção é “desviar” das barras, o que as move horizontalmente e reduz sua largura. Isso garante que não haja sobreposições e que as barras permaneçam comparáveis em termos de altura. Mas só funciona bem quando a variável categórica tem um pequeno número de níveis:

````python
sns.displot(
    penguins, 
    x='flipper_length_mm', 
    hue='sex', 
    multiple='dodge'
)
````

Como `displot()` é uma função em nível de figura e é desenhada em um `FacetGrid`, também é possível desenhar cada distribuição individual em um subtrama separado, atribuindo a segunda variável a col ou row em vez de (ou além de ) `hue`. Isso representa bem a distribuição de cada subconjunto, mas torna mais difícil fazer comparações diretas:

````python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    col='sex'
)
````

Nenhuma dessas abordagens é perfeita e em breve veremos algumas alternativas para um histograma que são mais adequadas para a tarefa de comparação.
